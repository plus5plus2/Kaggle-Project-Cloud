```python
def plot25(sample_x,th):
    sample_x[sample_x < th] = 0
    x_reshape = sample_x.reshape(-1, 64, 64) # reshape 

    fig = plt.figure(figsize =(15,15))
    for i in range(0,25):
        plt.subplot(5, 5, i+1)
        plt.imshow(x_reshape[i],cmap='gray')
    
    plt.tight_layout()
    print '----',th,'----'
    plt.show()
    fig.savefig(str(th)+'-thr.png')
    return plt
```

```python
plot25(sample_x,0)
```

```python
th_can=np.arange(125,256,10)
for th in th_can:
    p=plot25(sample_x,th)
```
